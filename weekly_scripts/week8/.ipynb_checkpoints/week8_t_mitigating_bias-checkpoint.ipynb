This notebook considers the problem of predicting recidivism -- will an individual arrested for a crime be arrested again? 
ProPublica considered the problems of predicting recidivism when they analyzed a tool called **COMPAS** that is used for this purpose. They obtained and analyzed information on 11,757 defendants in Broward Country, Florida. They found differences in how, for example, African-Americans are scored.

## 1. Build a Baseline Model

```python
from sklearn.model_selection import train_test_split

X = compas_pd.drop(columns=["is_recid"]) # remove target value
y = compas_pd["is_recid"]
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
from hyperopt import fmin, hp, tpe, Trials, STATUS_OK
import mlflow
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, recall_score
from xgboost import XGBClassifier
from xgboost.callback import EarlyStopping

mlflow.xgboost.autolog(disable=False, log_models=True)

def predict_xgboost(X_train_xgb, y_train_xgb, X_val_xgb, y_val_xgb, params):
    mlflow.xgboost.autolog(disable=False, log_models=True)

    pos_weight = (len(y_train_xgb) / y_train_xgb.sum()) - 1

    model = XGBClassifier(
        use_label_encoder=False,
        n_jobs=4,
        n_estimators=2000,
        random_state=0,
        scale_pos_weight=pos_weight,
        max_depth=int(params['max_depth']),
        learning_rate=params['learning_rate'],
        min_child_weight=params['min_child_weight'],
        subsample=params['subsample'],
        colsample_bytree=params['colsample_bytree'],
        eval_metric="logloss"
    )

    model.fit(
        X_train_xgb, y_train_xgb,
        eval_set=[(X_val_xgb, y_val_xgb)],
        # callbacks=[EarlyStopping(rounds=10, save_best=True)],
        verbose=False
    )
    return model.predict(X_val_xgb)

def run_data_experiment(data_experiment_tag, drop_cols=[], max_evals=50):
    def train_model(params):
        mlflow.set_tag("data_experiment", data_experiment_tag)
        X_train_mod = X_train.drop(columns=drop_cols, errors="ignore")
        X_val_mod = X_val.drop(columns=drop_cols, errors="ignore")

        y_pred = predict_xgboost(X_train_mod, y_train, X_val_mod, y_val, params)
        f1 = f1_score(y_val, y_pred)
        mlflow.log_metrics({
            'accuracy': accuracy_score(y_val, y_pred),
            'f1_score': f1,
            'recall': recall_score(y_val, y_pred)
        })
        return {'status': STATUS_OK, 'loss': -f1, 'params': params}

    search_space = {
        'max_depth': hp.quniform('max_depth', 2, 10, 1),
        'learning_rate': hp.loguniform('learning_rate', np.log(0.0001), np.log(0.1)),
        'min_child_weight': hp.qloguniform('min_child_weight', np.log(1), np.log(10), 1),
        'subsample': hp.uniform('subsample', 0.5, 1),
        'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1)
    }

    trials = Trials()
    best_params = fmin(
        fn=train_model,
        space=search_space,
        algo=tpe.suggest,
        max_evals=max_evals,
        trials=trials
    )
    return best_params, trials

def train_and_save_final_model(best_params, X_train, y_train, X_val, y_val, drop_cols=[], model_path="final_xgb_model.pkl"):
    """
    Train final XGBoost model on full training + validation data using best hyperparameters,
    drop specified columns, and save the trained model to disk.

    Args:
        best_params (dict): Best hyperparameters with keys like max_depth, learning_rate, etc.
        X_train (pd.DataFrame): Training features
        y_train (pd.Series): Training labels
        X_val (pd.DataFrame): Validation features
        y_val (pd.Series): Validation labels
        drop_cols (list): Columns to drop from features (e.g., demographic columns)
        model_path (str): Path to save the final model file

    Returns:
        XGBClassifier: The trained final model
    """
    # Fix parameter types
    best_params_fixed = {
        'max_depth': int(best_params['max_depth']),
        'learning_rate': best_params['learning_rate'],
        'min_child_weight': int(best_params['min_child_weight']),
        'subsample': best_params['subsample'],
        'colsample_bytree': best_params['colsample_bytree']
    }

    # Combine train and validation sets
    X_full = pd.concat([X_train, X_val], ignore_index=True)
    y_full = pd.concat([y_train, y_val], ignore_index=True)

    # Compute scale_pos_weight for imbalanced data
    pos_weight = (len(y_full) / y_full.sum()) - 1

    # Initialize and train model
    final_model = XGBClassifier(
        use_label_encoder=False,
        n_jobs=4,
        n_estimators=2000,
        random_state=0,
        scale_pos_weight=pos_weight,
        eval_metric="logloss",
        **best_params_fixed
    )

    final_model.fit(
        X_full.drop(columns=drop_cols, errors="ignore"),
        y_full,
        verbose=True
    )

    # Save model to disk
    joblib.dump(final_model, model_path)
    print(f"Final model saved to {model_path}")

    return final_model
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, ConfusionMatrixDisplay
from fairlearn.metrics import MetricFrame
import mlflow.pyfunc

# You need to implement or import these functions for fairness metrics
def false_positive_rate(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    fp = cm[0,1]
    tn = cm[0,0]
    return fp / (fp + tn) if (fp + tn) > 0 else 0

def false_negative_rate(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    fn = cm[1,0]
    tp = cm[1,1]
    return fn / (fn + tp) if (fn + tp) > 0 else 0

def selection_rate(y_true, y_pred):
    return sum(y_pred) / len(y_pred) if len(y_pred) > 0 else 0

def show_confusion_matrix_diff(model_path, X_val, y_val, drop_cols=[], race="African_American", 
                               sensitive_column="race_African_American", fair_used=False
                               ):
    """
    Show confusion matrices comparison for a given saved model loaded directly (joblib or sklearn).
    """
    model = joblib.load(model_path)

    if fair_used==False:
        X_val_mod = X_val.drop(columns=drop_cols, errors="ignore")
        recid_df = pd.concat([X_val.reset_index(drop=True), y_val.reset_index(drop=True)], axis=1)
        recid_df["prediction"] = model.predict(X_val_mod).astype('int32')
        
    elif fair_used==True:
        y_pred = predict_fair_model(model, X_val, sensitive_column, drop_cols)
        recid_df = pd.concat([X_val.reset_index(drop=True), y_val.reset_index(drop=True)], axis=1)
        recid_df["prediction"] = y_pred.astype('int32')

    def get_cm(label):
        data_pd = recid_df[recid_df[f"race_{race}"] == label][["is_recid", "prediction"]]
        return confusion_matrix(data_pd['is_recid'], data_pd['prediction'], normalize='all')

    not_race_cm = get_cm(0)
    race_cm = get_cm(1)

    _, axes = plt.subplots(1, 3, figsize=(20, 5), sharey='row')
    plt.rcParams.update({'font.size': 16})
    labels = ["Not Recid", "Recid"]
    ConfusionMatrixDisplay(not_race_cm, display_labels=labels).plot(ax=axes[0], cmap='Blues', colorbar=False).ax_.set_title(f"Not race_{race}")
    ConfusionMatrixDisplay(race_cm, display_labels=labels).plot(ax=axes[1], cmap='Blues', colorbar=False).ax_.set_title(f"race_{race}")
    ConfusionMatrixDisplay(race_cm - not_race_cm, display_labels=labels).plot(ax=axes[2], cmap='Purples', colorbar=False).ax_.set_title("Difference")
    plt.show()

def show_fairness_metrics(model_path, drop_cols=[], fair_used=False):
    """
    Compute and display fairness-related metrics by race group using the best model
    loaded from disk. Returns a MetricFrame with the breakdown by group.
    """
    # Load the saved model
    model = joblib.load(model_path)

    if fair_used==False:
        X_val_mod = X_val.drop(columns=drop_cols, errors="ignore")
        recid_df = pd.concat([X_val.reset_index(drop=True), y_val.reset_index(drop=True)], axis=1)
        recid_df["prediction"] = model.predict(X_val_mod).astype('int32')
        
    elif fair_used==True:
        y_pred = predict_fair_model(model, X_val, sensitive_column, drop_cols)
        recid_df = pd.concat([X_val.reset_index(drop=True), y_val.reset_index(drop=True)], axis=1)
        recid_df["prediction"] = y_pred.astype('int32')
        
    # Define fairness metrics
    metrics = {
        "accuracy": accuracy_score,
        "true positive(Recall)": recall_score,
        "false positive rate(FPR)": false_positive_rate,
        "false negative rate(FNR)": false_negative_rate,
        "selection rate": selection_rate
    }

    # Compute metrics by race group (e.g., race_African_American)
    mf = MetricFrame(
        metrics=metrics,
        y_true=recid_df["is_recid"],
        y_pred=recid_df["prediction"],
        sensitive_features=recid_df["race_African_American"],
        control_features=recid_df["is_recid"]
    ).by_group

    # print(mf)
    return mf
```

## 3. Prioritize Fairness Metrics with Microsoft Fairlearn
- Equalizing fairness metrics removes discrepancies but may reduce accuracy.
- Fairlearn offers tools to analyze and mitigate fairness gaps.
- Mitigation can be done by preprocessing (training) or postprocessing (adjusting outputs).
- Binary classifiers use thresholds to decide predictions; these can vary by group.
- Fairlearn’s ThresholdOptimizer sets group-specific thresholds to achieve equalized odds.

```python
from fairlearn.postprocessing import ThresholdOptimizer
import joblib

def train_fair_final_model(best_params, X_train, y_train, X_val, y_val,
                           sensitive_column, constraint="equalized_odds", model_path="fair_model.pkl", drop_cols=[]):
    """
    Train a fairness-aware XGBoost model using Fairlearn's ThresholdOptimizer and save to disk.
    """
    # Fix parameter types
    best_params_fixed = {
        'max_depth': int(best_params['max_depth']),
        'learning_rate': best_params['learning_rate'],
        'min_child_weight': int(best_params['min_child_weight']),
        'subsample': best_params['subsample'],
        'colsample_bytree': best_params['colsample_bytree']
    }

    # Merge train and validation sets
    X_full = pd.concat([X_train, X_val], ignore_index=True)
    y_full = pd.concat([y_train, y_val], ignore_index=True)

    # Compute positive class weight to address imbalance
    pos_weight = (len(y_full) / y_full.sum()) - 1

    # Train base model
    base_model = XGBClassifier(
        use_label_encoder=False,
        n_jobs=4,
        n_estimators=2000,
        random_state=0,
        scale_pos_weight=pos_weight,
        eval_metric="logloss",
        **best_params_fixed
    )

    # Prepare features
    features = X_full.drop(columns=drop_cols + [sensitive_column], errors="ignore")
    sensitive_features = X_full[sensitive_column]

    # Fit fairness post-processor
    optimizer = ThresholdOptimizer(
        estimator=base_model,
        constraints=constraint,
        objective="accuracy_score",
        predict_method="predict_proba"
    )
    optimizer.fit(features, y_full, sensitive_features=sensitive_features)

    # Save the model
    joblib.dump(optimizer, model_path)
    print(f"Fairness-aware model saved to {model_path}")

    return optimizer

def predict_fair_model(model, X, sensitive_column, drop_cols=[]):
    """
    Make predictions with a trained Fairlearn ThresholdOptimizer model.
    """
    features = X.drop(columns=drop_cols + [sensitive_column], errors="ignore")
    sensitive_features = X[sensitive_column]
    return model.predict(features, sensitive_features=sensitive_features, random_state=42)
```

```python
# 1. finding best model superparameters
drop_cols = [c for c in X_train.columns if c.startswith("sex") or c.startswith("race")]
sensitive_column = "race_African_American"
best_params, trials = run_data_experiment(data_experiment_tag="fairness_run", drop_cols=drop_cols)

# 2. Train model with fairness calibiration
fair_model = train_fair_final_model(
    best_params,
    X_train,
    y_train,
    X_val,
    y_val,
    sensitive_column=sensitive_column,
    constraint="equalized_odds",
    drop_cols=drop_cols, 
    model_path="fair_model_eqodds.pkl"
)
```

## 4. Tradeoff between fairness metrics and prediction accuracy

```python
import seaborn as sns

def compare_models_metrics_heatmap(
    model_paths,
    model_names,
    fair_used_flags,
    drop_cols_list,
    sensitive_column="race_African_American"
):
    summary = []

    for model_path, model_name, fair_used, drop_cols in zip(model_paths, model_names, fair_used_flags, drop_cols_list):
        mf = show_fairness_metrics(model_path, drop_cols=drop_cols, fair_used=fair_used)

        acc_0 = mf.loc[(0, 0), "accuracy"]
        acc_1 = mf.loc[(1, 0), "accuracy"]
        recall_gap = mf.loc[:, "true positive(Recall)"].max() - mf.loc[:, "true positive(Recall)"].min()
        fpr_gap = mf.loc[:, "false positive rate(FPR)"].max() - mf.loc[:, "false positive rate(FPR)"].min()
        dp_gap = mf.loc[:, "selection rate"].max() - mf.loc[:, "selection rate"].min()

        summary.append({
            "model": model_name,
            "recall_gap": recall_gap,
            "FPR_gap": fpr_gap,
            "demographic_parity_gap": dp_gap,
            "accuracy_not_recid": acc_0,
            "accuracy_recid": acc_1
        })

    df_summary = pd.DataFrame(summary).set_index("model")
    df_fairness = df_summary[["recall_gap", "FPR_gap", "demographic_parity_gap"]]
    df_accuracy = df_summary[["accuracy_not_recid", "accuracy_recid"]]

    # Create subplots for the two heatmaps
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))

    # Fairness gaps heatmap
    sns.heatmap(
        df_fairness,
        annot=True,
        fmt=".3f",
        cmap="RdYlGn_r",
        center=df_fairness.values.mean(),
        cbar_kws={'label': 'Gap Value'},
        ax=axes[0]
    )
    axes[0].set_title("Fairness Gaps by Model")
    axes[0].set_xlabel("Fairness Metric")
    axes[0].set_ylabel("Model")
    axes[0].tick_params(axis='x', rotation=30)

    # Accuracy per group heatmap
    sns.heatmap(
        df_accuracy,
        annot=True,
        fmt=".3f",
        cmap="RdYlGn_r",
        center=df_accuracy.values.mean(),
        cbar_kws={'label': 'Accuracy'},
        ax=axes[1],
        yticklabels=False
    )
    axes[1].set_title("Group Accuracy by Model")
    axes[1].set_xlabel("Group")
    # axes[1].set_ylabel("")
    axes[1].tick_params(axis='x', rotation=30)

    plt.tight_layout()
    plt.show()

    return df_summary
```

```python
drop_cols = [c for c in X_train.columns if c.startswith("sex") or c.startswith("race")]

model_paths = [
    "final_xgb_model.pkl",
    "final_xgb_model_nodemo.pkl",
    "fair_model_eqodds.pkl"
]

model_names = [
    "with Demographic var",
    "No Demographic var",
    "with Fairlearn mod"
]

fair_used_flags = [False, False, True]

drop_cols_list = [
    [],             
    drop_cols,      
    drop_cols       
]

df_result = compare_models_metrics_heatmap(
    model_paths=model_paths,
    model_names=model_names,
    fair_used_flags=fair_used_flags,
    drop_cols_list=drop_cols_list,
    sensitive_column="race_African_American"
)
```
