## 1. Hello World!

```python
print("Hello, world!")  # This prints text to the screen
```

## 2. What is a Variable?

```python
# 3. What is a Variable?

# Think of a variable as a label or a box that stores a value.
# The = sign assigns the value on the right to the name on the left.

x = 5  # We are assigning the number 5 to the variable named x
print("x is:", x)

my_name = "Alice"  # Assigning a string to a variable
print("My name is:", my_name)

# You can use variables to store numbers, text, or other data types
```

## 3. The Types of Variables

```python
# Let's create some variables
my_number = 10         # an integer
my_float = 3.14        # a floating-point number
my_text = "AI is fun!" # a string
is_sunny = True        # a boolean (True or False)

# Let's print them out
print("my_number:", my_number, type(my_number))
print("my_float:", my_float, type(my_float))
print("my_text:", my_text, type(my_text))
print("is_sunny:", is_sunny, type(is_sunny))
```

###  1) Math Operations

```python
a = 7
b = 3

print("Addition: a + b =", a + b)
print("Subtraction: a - b =", a - b)
print("Multiplication: a * b =", a * b)
print("Division: a / b =", a / b)
print("Power: a ** b =", a ** b)
print("Modulus (remainder): a % b =", a % b)
```

###  2) Strings and Combining Text

```python
# String variables can be easily combined with "+"
name = "Alice"
greeting = "Hello, " + name + "!"
print(greeting)

# f-strings (formatted strings) let you insert variables into text easily
age = 20
message = f"{name} is {age} years old."
print(message)
```

###  3) Booleans and Comparison

```python
x = 5
y = 10

# These expressions compare values and return True or False
print("x == y:", x == y)   # equal to?
print("x != y:", x != y)   # not equal to?
print("x > y:", x > y)
print("x < y:", x < y)
print("x >= y:", x >= y)
print("x <= y:", x <= y)
```

## 4. If-Else Statements

```python
# If-elif-else lets you make decisions in your program

temperature = 25

if temperature > 30:
    print("It's hot outside.")
elif temperature > 20:
    print("It's a nice day.")
else:
    print("It's a bit cold.")
```

## 5. Functions

```python
# Functions let you group code into reusable blocks

# Define a function that greets a person
def greet(name):
    print(f"Hello, {name}!")

# Define a function that adds two numbers and returns the result
def add_numbers(a, b):
    return a + b

# Call the greet function
greet("Alice")

# Call the add_numbers function and store the result
result = add_numbers(5, 7)
print("5 + 7 =", result)
```
